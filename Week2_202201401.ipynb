```python
#Write Code for Logistic Regression here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from google.colab import files
uploaded = files.upload()


cancer_dataset = pd.read_csv('dataR2.csv')

x = cancer_dataset[['Age', 'BMI', 'Glucose', 'Insulin', 'HOMA', 'Leptin', 'Adiponectin', 'Resistin', 'MCP.1']]
y = cancer_dataset['Classification']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

log_reg = LogisticRegression(max_iter=10000)
log_reg.fit(x_train, y_train)

print(log_reg.score(x_test,y_test))
```

SVM(Support Vector Machines)

```python
#Write Code for SVM here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

from google.colab import files
uploaded = files.upload()


cancer_dataset = pd.read_csv('dataR2.csv')

x = cancer_dataset[['Age', 'BMI', 'Glucose', 'Insulin', 'HOMA', 'Leptin', 'Adiponectin', 'Resistin', 'MCP.1']]
y = cancer_dataset['Classification']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

clf=SVC(kernel='linear',C=3)
clf.fit(x_train,y_train)

print(clf.score(x_test,y_test))
```

```python
#Code for Linear Regression here

#Code for Linear Regression here

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from google.colab import files
uploaded = files.upload()

real_estate_dataset = pd.read_excel('Real estate valuation data set.xlsx')

columns_to_remove = ['No', 'X1 transaction date']
real_estate_dataset = real_estate_dataset.drop(columns=columns_to_remove)


X = real_estate_dataset[['X2 house age', 'X3 distance to the nearest MRT station', 'X4 number of convenience stores', 'X5 latitude', 'X6 longitude']]
y = real_estate_dataset['Y house price of unit area']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

print(model.score(X_test,y_test))
```
